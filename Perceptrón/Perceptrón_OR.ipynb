## Perceptrón

### Función OR

El código implementa el modelo básico del perceptrón para realizar una clasificación binaria y sigue la siguiente serie de pasos:


1.   Inicialización

    Entradas (X): [[1,1], [0,1], [1,0], [0,0]] (Las 4 combinaciones posibles).

    Salidas Reales (Y): [1, 1, 1, 0] (La tabla de verdad de la función OR).

        Parámetros Iniciales:

            pesos = [5, 3]

            sesgo = 3

            tasa_de_aprendizaje = 0.1

2.   Predicción y Activación:

Calcula la suma ponderada (z) para un dato: $z=x_1 w_1 +x_2 w_2+b.$

Aplica la función escalón (umbral de 0): Si z≥0, la predicción ($y_{pred}$) es 1; si no, es 0.

3.   Entrenamiento:

Itera a través de las `epocas` (en este caso 15).

```python
"""
Definimos las entradas y salidas del perceptrón, junto con las variables a usar en el entrenamiento
"""

X = [[1,1],[0,1],[1,0],[0,0]]
Y = [1,1,1,0]
sesgo = 3
pesos = [5,3]
tasa_de_aprendizaje = 0.1
```

```python
"""
Definimos la clase del perceptrón
"""
class Perceptron:
    def __init__ (self,pesos, sesgo, tasa_de_aprendizaje):


      """
        se documenta la función
        parametros:
           pesos: Es una lista
           sesgo: Es un flotante
           tasa_de_aprendizaje: Es un flotante
      """
      self.pesos = pesos
      self.sesgo = sesgo
      self.tasa_de_aprendizaje = tasa_de_aprendizaje
       # aquí ya acabé de configurar la clase perceptrón
       # comienzo a definir la función de activación
    def funcion_de_activacion(self, z):
        """"
            Parámetros de la función de activación:
            z es el producto de X(pesos+sesgo)
        """
        if z >= 0:
          return 1
        else:
          return 0
        # Fin de la función de activación

        # inicio de la función de entremanient
    def entrenamiento(self, X, Y, epocas):
        """

        """
       # X = [[1,1],[1,0],[0,1],[0,0]]
      # Y = [1,1,1,0]
        for epoca in range(epocas):
          for dato in X:
            # Definimos los datos de la función de entranamiento, que son necesarios para los cálculos
            x1 = dato[0]
            x2 = dato[1]
            y_real = Y[X.index(dato)]
            w1 = self.pesos[0]
            w2 = self.pesos[1]
            b = self.sesgo

            #Hacemos la predicción del modelo
            z = x1*w1 + x2*w2 + b
            y_pred = self.funcion_de_activacion(z)
            # Actualizamos los pesos
            self.pesos[0] = w1 + self.tasa_de_aprendizaje * (y_real - y_pred) * x1
            self.pesos[1] = w2 + self.tasa_de_aprendizaje * (y_real - y_pred) * x2
            self.sesgo = b + self.tasa_de_aprendizaje * (y_real - y_pred)
            print(f"Dato {dato}, Pesos: {self.pesos} y Sesgo: {self.sesgo}")
          print(f"Epoca {epoca}, Pesos: {self.pesos} y Sesgo: {self.sesgo}")
          # fin del entrenamiento

        # inicio de la predicción
        def prediccion(sef, X, Y):
          for dato in X:
            # Definimos los datos de la función de entranamiento, que son necesarios para los cálculos
            x1 = dato[0]
            x2 = dato[1]
            y_real = Y[X.index(dato)]
            w1 = self.pesos[0]
            w2 = self.pesos[1]
            b = self.sesgo
          #Hacemos la predicción del modelo
            z = x1*w1 + x2*w2 + b
            y_pred = self.funcion_de_activacion(z)
            print(f"Dato {dato}, prediccion {y_pred}, valor real {y_real}")


```
