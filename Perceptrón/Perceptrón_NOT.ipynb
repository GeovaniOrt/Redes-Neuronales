## Perceptrón

### Función NOT

El código implementa el modelo básico del perceptrón para realizar una clasificación binaria para realizar la operación lógica `NOT`y sigue la siguiente serie de pasos:


1.   Inicialización

    Entradas (X): [1,0] (Para NOT, solo existen dos posibles entradas).

    Salidas Reales (Y): [0, 1] (La función NOT, niega (invierte) las cosas).

    Parámetros Iniciales:

          peso = [-1]

          sesgo = 2

          tasa_de_aprendizaje = 0.5

2.   Predicción y Activación:

Calcula la suma ponderada (z) para un dato: $z=x w +b.$

Aplica la función escalón (umbral de 0): Si z≥0, la predicción ($y_{pred}$) es 1; si no, es 0.

3.   Entrenamiento:

Itera a través de las `epocas` (en este caso 150, épocas).


```python
"""
Definimos las entradas y salidas del perceptrón, junto con las variables a usar en el entrenamiento
"""

X = [[1], [0]]
Y = [0, 1]
sesgo = 2
peso = [-1]
tasa_de_aprendizaje = 0.5
```

```python
"""
Definimos la clase del perceptrón
"""
class Perceptron:
    def __init__ (self,peso, sesgo, tasa_de_aprendizaje):


      """
        se documenta la función
        parametros:
           peso: Es una lista
           sesgo: Es un flotante
           tasa_de_aprendizaje: Es un flotante
      """
      self.peso = peso
      self.sesgo = sesgo
      self.tasa_de_aprendizaje = tasa_de_aprendizaje
       # aquí ya acabé de configurar la clase perceptrón
       # comienzo a definir la función de activación
    def funcion_de_activacion(self, z):
        """"
            Parámetros de la función de activación:
            z es el producto de X(peso+sesgo)
        """
        if z >= 0:
          return 1
        else:
          return 0
        # Fin de la función de activación

        # inicio de la función de entremaniento
    def entrenamiento(self, X, Y, epocas):
        """

        """
        for epoca in range(epocas):
          for dato in X:
            # Definimos los datos de la función de entranamiento, que son necesarios para los cálculos
            x = dato[0]
            y_real = Y[X.index(dato)]
            w = self.peso[0]
            b = self.sesgo

            #Hacemos la predicción del modelo
            z = x*w + b
            y_pred = self.funcion_de_activacion(z)
            # Actualizamos los peso
            self.peso[0] = w + self.tasa_de_aprendizaje * (y_real - y_pred) * x
            self.sesgo = b + self.tasa_de_aprendizaje * (y_real - y_pred)
            print(f"Dato {dato}, peso: {self.peso} y Sesgo: {self.sesgo}")
          print(f"Epoca {epoca}, peso: {self.peso} y Sesgo: {self.sesgo}")
          # fin del entrenamiento

        # inicio de la predicción
        def prediccion(sef, X, Y):
          for dato in X:
            # Definimos los datos de la función de entranamiento, que son necesarios para los cálculos
            x = dato[0]
            y_real = Y[X.index(dato)]
            w = self.peso[0]
            b = self.sesgo
          #Hacemos la predicción del modelo
            z = x*w + b
            y_pred = self.funcion_de_activacion(z)
            print(f"Dato {dato}, prediccion {y_pred}, valor real {y_real}")


```

```python
"""
Esta es la inicialización del perceptrón (la clase)
"""
p = Perceptron(peso, sesgo, tasa_de_aprendizaje)
```
