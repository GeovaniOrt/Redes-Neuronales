# Redes Nweuronales

## McCullochPitts

La neurona de McCulloch-Pitts (`MCP`) resuelve la operación lógica `OR` mediante el ajuste de su Umbral (`θ`) a un valor que permite la activación de la neurona si al menos una de sus entradas excitatorias está activa.

La neurona MCP resuelve la función `OR` porque esta es linealmente separable . Al configurar el umbral a1, solo se necesita una unidad de entrada (una entrada activa) para alcanzar el umbral y producir una salida positiva (1), replicando así la lógica de la compuerta `OR`.

### Función OR

```python
import numpy as np
```

```python
"""
Entrada de parametros/variables para la neurona
"""
# Entradas para la neurona
a = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
# Salidas
b = np.array([0, 1, 1, 1])
# Pesos para las entradas
c = [3, 2]
# Umbral (θ)
umbral = 2
```

```python
"""
Funcón para el funcionamiento de la neurona de McCulloch-Pitts
"""
class McCullochPitts:
    def __init__(self, weights, umbral):
        self.weights = weights
        self.umbral = umbral

    def predict(self, a):
        # Lista donde se guardarán las salidas
        predictions = []
        for i in range(a.shape[0]):
            weighted_sum = np.dot(self.weights, a[i])
            print(f"Input: {a[i]}, Weighted Sum: {weighted_sum}")
            if weighted_sum >= self.umbral:
                predictions.append(1)
            else:
                predictions.append(0)
        return predictions
```

```python
"""
Imprime y muestra resultados de la neurona
"""
mcp = McCullochPitts(c, umbral)
# Calculamos las predicciones
predictions = mcp.predict(a)
# Muestra el resultado final
print("Predictions:", predictions)
```
