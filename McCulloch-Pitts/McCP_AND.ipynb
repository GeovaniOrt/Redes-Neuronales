## McCullochPitts

Est谩 neurona es el primer modelo matem谩tico de una neurona artificial, propuesto en 1943 por el neurofisi贸logo Warren McCulloch y el l贸gico Walter Pitts.  Fue la piedra angular que sent贸 las bases para el desarrollo de las redes neuronales artificiales y el campo del Deep Learning.

### Funci贸n AND

La neurona de McCulloch-Pitts resuelve la operaci贸n l贸gica `AND` estableciendo un umbral (`胃`) que solo es superado cuando TODAS las entradas binarias son activas (igual a 1).

De esta manera, la neurona solo se "activa" (salida = 1) cuando todas sus entradas est谩n activas , que es la definici贸n de la operaci贸n `AND`.

```python
"""
Entrada de parametros/variables para la neurona
"""
import numpy as np
# Entradas para la neurona
x = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
# Salidas
y = np.array([0, 0, 0, 1])
# Pesos para las entradas
z = [1, 1]
# Umbral (胃)
umbral = 2
```

```python
"""
Func贸n para el funcionamiento de la neurona de McCulloch-Pitts
"""
class McCullochPitts:
    def __init__(self, weights, umbral):
        self.weights = weights
        self.umbral = umbral

    def predict(self, x):
        # Lista donde se guardar谩n las salidas
        predictions = []
        for i in range(x.shape[0]):
            weighted_sum = np.dot(self.weights, x[i])
            print(f"Input: {x[i]}, Weighted Sum: {weighted_sum}")
            if weighted_sum >= self.umbral:
                predictions.append(1)
            else:
                predictions.append(0)
        return predictions
```

```python
"""
Imprime y muestra resultados de la neurona
"""
mcp = McCullochPitts(z, umbral)
# Calculamos las predicciones
predictions = mcp.predict(x)
# Muestra el resultado final
print("Predictions:", predictions)
```
