# Redes Neuronales

## McCullochPitts

### Función NOT

La neurona de McCulloch-Pitts (MCP) resuelve la operación lógica `NOT` utilizando una combinación de una ***entrada excitatoria permanente*** y una ***entrada inhibidora*** conectada a la variable que se quiere negar. El truco está en establecer un umbral (a diferencia de AND y OR, solo se necesita un peso(w)) que solo permita la activación si la entrada inhibidora está apagada (0).

Al usar la entrada inhibidora absoluta conectada a x, y una entrada excitatoria fija para establecer el estado predeterminado de 'ON' (Salida=1), el modelo MCP logra la negación lógica.

```python
import numpy as np
```

```python
"""
Entrada de parametros/variables para la neurona
"""
# Entradas para la neurona
x = np.array([
    [0],
    [1]
])
# Salidas
y = np.array([0, 1])
# Pesos para las entradas
w = [-1]
# Umbral (θ)
umbral = 0
```

```python
"""
Funcón para el funcionamiento de la neurona de McCulloch-Pitts
"""
class McCullochPitts:
    def __init__(self, weights, umbral):
        self.weights = weights
        self.umbral = umbral

    def predict(self, a):
        # Lista donde se guardarán las salidas
        predictions = []
        for i in range(x.shape[0]):
            weighted_sum = np.dot(self.weights, x[i])
            print(f"Input: {x[i]}, Weighted Sum: {weighted_sum}")
            if weighted_sum >= self.umbral:
                predictions.append(1)
            else:
                predictions.append(0)
        return predictions
```

```python
"""
Imprime y muestra resultados de la neurona
"""
mcp = McCullochPitts(w, umbral)
# Calculamos las predicciones
predictions = mcp.predict(y)
# Muestra el resultado final
print("Predictions:", predictions)
```
